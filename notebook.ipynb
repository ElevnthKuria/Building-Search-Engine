```python
import pandas as pd

questions_file = 'stacksample/Questions.csv'
answers_file = 'stacksample/Answers.csv'
subset_size = 200000

# Read the questions.csv file into a DataFrame
questions_df = pd.read_csv(questions_file,encoding='latin-1')

# Read the answers.csv file into a DataFrame
answers_df = pd.read_csv(answers_file,encoding='latin-1')

# Set 'Id' as the index for both DataFrames
questions_df.set_index('Id', inplace=True)
answers_df.set_index('Id', inplace=True)

# Merge the questions and answers DataFrames on the index
merged_df = questions_df.merge(answers_df, left_on='Id', right_on='ParentId',suffixes=('_questions', '_answers'))
```

```python
# URL of the model
module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"

# Download the model
loaded_model = hub.load(module_url)

#new column title_body
questions['title_body'] = questions['Title'] + '' + questions['Body_questions']

# Applying the model on tilte_body column and store the result in new column
questions['Embeddings'] = questions['title_body'].apply(lambda x: loaded_model([x])[0].numpy())

# Convert embedding column to a numpy array
embeddings = np.array(questions['Embeddings'].tolist())
```

```python
import nmslib
import numpy as np
import pandas as pd

# Create an nmslib index
index = nmslib.init(method='hnsw', space='cosinesimil')

# Add data points to the index. we computed embeddings in above block.
index.addDataPointBatch(embeddings)

# Build the index
index.createIndex({'post': 2},print_progress=True)

#test search query
a = 'How to install pip. i have a python virtual env and i want to install pip in it.'

#a function that takes in text query and returns the USE embedding for it
#This embedding will later be used for search
def query2emb_func(text):
  #example_sentence = "Hello, how are you?"
  embeddings = loaded_model([text])
  return embeddings.numpy()

#embed the query
#it will return a <class 'tensorflow.python.framework.ops.EagerTensor'>
b = query2emb_func(a)
print(type(b))

#covert the tensor into a numpy array
#it will return <class 'numpy.ndarray'> type
b = np.array(b)
#print(type(b))

#check the shape
#it will be (1, 512) shape vector
b.shape
```

```python
#now we will search our embedding "b" in the index
#K nearest neighbour is used here
#we are finding the indices and distances of 6 nearest neigbours in our index
idxs, dists = index.knnQuery(b, k=6)

#now we will take above indices to print the data points from our questions DF
# we are returning Title, doc_id and Parent_id from the DF
for idx, dist in zip(idxs, dists):
  Title = questions.iloc[idx].Title
  doc_id = questions.iloc[idx].doc_id
  ParentId = questions.iloc[idx].ParentId
  #url = ref_df.iloc[idx].url
  print(f'cosine dist:{dist:.4f}')
  print(doc_id)
  print(Title)
  print(ParentId)
  print(f'\n---------------\n')
```

```python
from whoosh.fields import Schema, TEXT, ID
from whoosh.index import create_in
from whoosh.qparser import QueryParser
from whoosh.index import open_dir
from whoosh.qparser import QueryParser
from whoosh.scoring import TF_IDF
from whoosh import scoring

# Defining index schema
schema = Schema(doc_id=ID(stored=True), Title=TEXT(stored=True), title_body=TEXT(stored=True))

# Creating index directory
index_dir = "index_directory_1"

# Creating the index
ix = create_in(index_dir, schema)

# Open the index writer
writer = ix.writer()

# Iterate over the DataFrame and add documents to the index
# we have indexed title, title_body and doc_id
for index, row in questions.iterrows():
    writer.add_document(doc_id=str(row['doc_id']), Title = row['Title'] ,title_body=row['title_body'])

# Commit and close the writer
writer.commit()
```

```python
# keyword search on Title. 
# scoring will be tf-idf.

from whoosh.qparser import QueryParser

# create the query parser
qp = QueryParser("Title", schema=schema)

# parse the query
query_sentence = "How to install pip"
query = qp.parse(query_sentence)

# create a searcher object
searcher_tfidf = ix.searcher(weighting=scoring.TF_IDF())

# search documents and store them
# we are returing top 7 documents
results_tfidf = searcher_tfidf.search(query, limit=7, scored=True)

# print the documents
for hit in results_tfidf:
    print(hit["doc_id"])
    print('\n')
    print(hit["Title"])
    print('\n')
    print('------------------\n')
```
